```python
from google.colab import files
from pdf2image import convert_from_path
from transformers import pipeline
from PIL import Image
import pytesseract

# Optional but safe
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# Upload the PDF
uploaded = files.upload()
pdf_path = next(iter(uploaded))

# Convert to images
images = convert_from_path(pdf_path)
print(f"✅ Converted {len(images)} page(s) from PDF.")
```

```python
qa_pipeline = pipeline("document-question-answering", model="impira/layoutlm-document-qa")
```

```python
question = "What is the invoice number?"

all_results = []
for i, img in enumerate(images):
    answers = qa_pipeline(img, question=question)
    all_results.append({
        "page": i + 1,
        "answers": answers
    })
```

```python
for page_data in all_results:
    print(f"\n📄 Page {page_data['page']}:")
    for ans in page_data["answers"]:
        print(f"  ➤ Answer: {ans['answer']} (Score: {ans['score']:.2f})")
```
